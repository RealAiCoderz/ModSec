```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
regions = pd.read_csv("regions.csv")
products = pd.read_csv("products.csv")
customers = pd.read_csv("customers.csv")
sales = pd.read_csv("sales.csv")

sales["date"] = pd.to_datetime(sales["date"])
```

### 1. Visualize the distribution of sales amounts using a histogram

```python
bw = [50, 100, 200, 300, 400, 500, 600, 700, 800, 1000]

fig,ax = plt.subplots(len(bw), figsize=(10,20))
for i in range(len(bw)):
    ax[i].hist(sales['sales_amt'], bins=np.arange(min(sales['sales_amt']),max(sales['sales_amt']) + bw[i], bw[i]),
        edgecolor='black', linewidth=2)
plt.show()
```

### 2. Create a boxplot to identify any outliers in the sales data.

```python
fig, ax = plt.subplots()
ax.boxplot(sales.sales_amt, patch_artist=True)
plt.show()
```

```python
sales.boxplot(column=['sales_amt'], by='region_id', grid=True,
              color={'boxes':'grey', 'whiskers':'black', 'medians':'black'},
             patch_artist=True)
plt.show()
```

### 3. Explore the relationship between sales and other variables using scater plots.

```python
y = sales.sales_amt
x = [x.strftime("%b") for x in sales.date]

fig, ax = plt.subplots()
ax.scatter(x,y, alpha=0.2)
plt.show()
```

```python
y = sales.sales_amt
x = sales.region_id

fig, ax = plt.subplots()
ax.scatter(x,y, alpha=0.2)
plt.show()
```
