### 4. Normalize the 'sales_amount' column between 0 and 1

```python
minimum = sales.sales_amount.min()
maximum = sales.sales_amount.max()
average = sales.sales_amount.mean()
sales['sales_amount_normalized'] = (sales.sales_amount- minimum)/(maximum)
sales
```

### 5. Identify and remove outliers in Sales table

```python
# we will demonstrate this on 'sales_amount' column in the Sales table

col = sales.sales_amount
q1 = col.quantile(0.25)
q3 = col.quantile(0.75)

iqr = q3 - q1

# number of outliers:
indices = col.between(q1-1.5*iqr, q1+1.5*iqr)
indices.sum()
```

```python
sales = sales.loc[indices]
sales
```

```python
# Group the sales data by product category and calculate the total sales amount for each category
category_sales = cleaned_sales.groupby('category')['sales_amount'].sum()
category_sales
```

```python
# Group the sales data by month and year and calculate the average sales amount for each month
sales_data=cleaned_sales.copy()
sales_data['date'] = pd.to_datetime(sales_data['date'])
sales_data['month'] = sales_data['date'].dt.to_period('M')
sales_data['year']=sales_data['date'].dt.to_period('Y')
monthly_average_sales = sales_data.groupby('month')['sales_amount'].mean()
yearly_average_sales = sales_data.groupby('year')['sales_amount'].mean()

```

```python
print(monthly_average_sales,yearly_average_sales)
```

```python
# Aggregate the sales data by region and calculate the total sales amount for each region
region_sales = sales_data.groupby('region_name')['sales_amount'].sum()
region_sales
```

```python
# Group the sales data by customer segment and calculate the average sales amount for each segment
sales_data['customer_segment'] = sales_data['customer_name'].apply(lambda x: x.split()[0])
segment_average_sales = sales_data.groupby('customer_segment')['sales_amount'].mean()



```
